# Optimization by Iterative Method.  

We find the value of $x$ that minimizes the function $f(x) = (x-2)^3(x-6)+4$, where $x\in\mathbb{R},$ using iterative methods.

##  First derivative of $f(x)$

**Implement f(x), f_prime(x)** functions that return the value of $f(x), f'(x)$ for all $x \in \mathbb{R}$.

```python
# Write code here


def f(x):
    
    
    return (x-2)**3 * (x-6) + 4

def f_prime(x):
    
    
    return 4 * (x-2) ** 2 * (x-5)

```

```python
# Do Not Modify!!

print(f'f(6) = {f(6)}')
print(f'f\'(4) = {f_prime(4)}')
```

## Gradient Descent Method

**Implement GD_update** function that updates the parameter estimate using **gradient descent method.**

- Input : 
1. x_old : old parameter estimate  ($x_{t-1}$)
2. learning_rate ($\gamma$)
- Output : 
1. x_new : updated paramter estimate($x_{t}$) 

```python
def GD_update(x_old, learning_rate):
    # Write code here
    x_new = x_old - learning_rate * f_prime(x_old)
    return x_new
```

```python
# Do not modify!!

lr = 0.1
print(f_prime(3))
print(GD_update(3, lr))
print(f_prime(GD_update(3,lr)))
print(GD_update(GD_update(3, lr), lr))
```

## Find the value $x^*$ using gradient descent method.

Implement **GD_optimize** function that satisfies the following conditions.
* Input:<br/>
1) x_zero: initial parameter estimate ($x_0$)<br/>
2) learning rate<br/>
3) threshold
* Output:<br/>
1) params (list that contains parameter estimates $x_0$, $x_1$, $\cdots$, where last element is $x^*$)<br/>
2) values (list that contains values of $f(x_0)$, $f(x_1)$, $\cdots$, where last element is the minimum value, $f(x^*)$)

* How to Implement
1. We want to find the minimizer of $f(x)$ using gradient descent method.
2. Initialize with any $x_0 \in \mathbb{R}.$
3. For each step, <br/>
1) Update $x_{t-1}$ to $x_{t}$ using gradient descent method.<br/>
2) Print the $x_{t}$ and $f(x_{t})$ <br/>
3) Append the $x_{t}$ to "params" list <br/>
4) Append the $f(x_{t})$ to "values" list
4. Repeat step 3 until $|x_{t}-x_{t-1}| <$ threshold. 
5. Print $x^*$ and $f(x^*)$ when $x_t$ has converged.


```python
def GD_optimize(x_zero, learning_rate, threshold):
    # Initialize
    params = [x_zero]
    values = [f(x_zero)]
    # Write code here
    t = 0
    print(f'x_{t} is {x_zero}, f(x_{t}) is {f(x_zero)}')
    x_old = params[-1]
    x_new = GD_update(x_old, learning_rate)
    t += 1
    print(f'x_{t} is {x_new}, f(x_{t}) is {f(x_new)}.')
    params.append(x_new)
    values.append(f(x_new))

    while abs(params[-1] - params[-2]) >= threshold:    
        x_old = params[-1]
        x_new = GD_update(x_old, learning_rate)
        t += 1
        print(f'x_{t} is {x_new}, f(x_{t}) is {f(x_new)}.')
        params.append(x_new)
        values.append(f(x_new))
    print(f'x_star is {params[-1]}, f(x_star) is {values[-1]}')


    return params, values
```

```python
## Do Not Modify!!
x_zero = 0
learning_rate = 0.01
threshold = 1e-3

params, values = GD_optimize(x_zero, learning_rate, threshold)
```

## Sigmoid
**Implement sigmoid** function.


$\pi(x_i) = \dfrac{exp(\beta_0+\beta_1x_i)}{1+exp(\beta_0+\beta_1x_i)}$

* Input:
1. x_i : $x_i$
2. beta_0 : $\beta_0$
3. beta_1 : $\beta_1$

* Output:
$\pi(x_i)$

```python
def sigmoid(x_i, beta_0, beta_1):
    # Write code here
    z = np.exp(beta_0 + beta_1 * x_i)
    return z / (1+z)
```

## log-Likelihood($\beta$)

* Logistic Regression Problem:<br/>
$y_i \sim Ber(\pi(x_i))$, $\pi(x_i) = \dfrac{exp(\beta_0+\beta_1x_i)}{1+exp(\beta_0+\beta_1x_i)}$
<br/><br/>
Hence, $y_i$ is a binary variable taking values 0 or 1 with $P(y_i=1|x_i) = \pi(x_i), P(y_i=0|x_i) = 1-\pi(x_i)$<br/> and probability mass function of $y_i$ is <br/>
$p(y_i|x_i) = \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$
<br/>
Suppose that $y_1, y_2, ... , y_N$ are independent. Then the joint probability mass function of $y_1, y_2, ... , y_N$ is <br/>
$p(y_1, ... , y_N|x_1, ... , x_N) = \prod_{i=1}^{N} \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$.<br/>
==> $Likelihood(\beta)=L(\beta)= \prod_{i=1}^{N} \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$<br/>
==> $Log-likelihood(\beta) = logL(\beta) = l(\beta)$






**Implement log_likelihood** function that calculates the log_likelihood($\beta$) using **sigmoid** function.

* Input: 
1. beta_0, beta_1 : $\beta_0, \beta_1$
2. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
3. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
* Output:
value of logLikelihood($\beta$), $l(\beta) \in \mathbb{R}$

```python
def log_likelihood(x, y, beta_0, beta_1):
    # Write code here
    Pi_x = [sigmoid(x_i, beta_0, beta_1) for x_i in x]
    temp = [y_i*np.log(Pi_x[i]) + (1-y_i)*np.log(1-Pi_x[i]) for i, y_i in enumerate(y)]
    log_likelihood = sum(temp)
    return log_likelihood
```

## Score function

Score function is first derivative(gradeint) of log-likelihood function. It defined as below.<br/>$S(\beta)\overset{\underset{\mathrm{def}}{}}{=} \dfrac{\partial l(\beta)}{\partial \beta} = \begin{bmatrix}
\dfrac{\partial l(\beta)}{\partial \beta_0} \\
\dfrac{\partial l(\beta)}{\partial \beta_1} \end{bmatrix} = \bigtriangledown l(\beta) \in \mathbb{R^{2\times1}}$

**Implement score_function** that calculate score function $S(\beta)$ vector.

* Input: 
1. beta_0, beta_1 : $\beta_0, \beta_1$
2. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
3. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
* Output:
numpy-array of Score function $S(\beta) \in \mathbb{R^{2\times1}}$

```python
def score_function(x, y, beta_0, beta_1):
    # Write code here
    Pi_x = [sigmoid(x_i, beta_0, beta_1) for x_i in x]
    temp1 = [y_i - Pi_x[i] for i, y_i in enumerate(y)]
    temp2 = [x_i * temp1[i] for i, x_i in enumerate(x)]
    element1, element2 = sum(temp1), sum(temp2)
    return np.array([[element1], [element2]])
```

## MLE of Logistic Regression parameters $\beta_0, \beta_1$.

**Implement MLE** function that estimates Logistic Regression parameters $\beta_0, \beta_1$ <br/> using **Gradient-Descent method** 

* Input:
1. x : $\mathbf{X} \in \mathbb{R}^{N\times1}$
2. y : $\mathbf{y} \in \mathbb{R}^{N\times1}$
3. epochs : number of iterations
4. lr: learning rate

* Output:
1. params (list that contains parameters $\beta^{(0)}$, $\beta^{(1)}$, $\cdots$, where last element is true parameter, $\beta^*$)
2. values (list that contains values of log-likelihood $l(\beta^{(0)})$, $l(\beta^{(1)})$, $\cdots$, where last element is  $l(\beta^*)$

* How to Implement
1. We want to find the minimizer of $-l(\beta)$
2. Initialize with $\beta^{(0)} = [0,0]^T \in \mathbb{R^{2 \times 1}}.$
3. For each epoch($k$), where $k$ $\in$ (0, 1, ... epochs-1)<br/>
1) Update $\beta^{(k)}$ to $\beta^{(k+1)}$ using Gradient-Descent method.<br/>
2) Print the $\beta^{(k+1)}$ and $l(\beta^{(k+1)})$ <br/>
3) Append the $\beta^{(k+1)}$ to "params" list <br/>
4) Append the $l(\beta^{(k+1)})$ to "values" list
5. Print $\beta^{(epochs)}$ and $l(\beta^{(epochs)})$.  (This should be close to $\beta^*$ and $l(\beta^*)$.)

```python
def MLE(x, y, epochs, lr=0.01):
    #Initialize
    params = []
    values = []
    beta_k = np.zeros(2).reshape(-1,1) # Initial beta_0, beta_1 = 0, 0
    # Write code here
    for epoch in range(epochs):
        params.append(beta_k)
        beta_0, beta_1 = beta_k[0,0], beta_k[1,0]
        value = log_likelihood(x,y,beta_0,beta_1)
        values.append(value)
        print(f'beta_0^({epoch}) is {beta_0}, beta_1^({epoch}) is {beta_1}, L(beta) is {value}')
        score = score_function(x, y, beta_0, beta_1)
        beta_k = beta_k + lr * score
    params.append(beta_k)
    values.append(value)    
    print(f'beta_0_star is {beta_0}, beta_1_star is {beta_1}, L(beta_star) is {values[-1]}') 

    return params, values    
```

```python
## DO NOT MODIFY!!

params, values = MLE(x, y, epochs = 15)
```
