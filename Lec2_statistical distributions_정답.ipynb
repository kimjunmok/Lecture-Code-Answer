```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

### Sampling from a distribution

1. Normal distribution

```python
mu, sigma=0, 2
```

```python
np.random.seed(3)
```

```python
np.random.normal(loc=mu, scale=sigma) #N(0,4)
```

```python
X=np.random.normal(loc=mu, scale=sigma, size=10000)
```

```python
len(X)
```

```python
np.mean(X)
```

```python
np.var(X)*len(X)/(len(X)-1)
```

```python
sum((X-np.mean(X))**2)/(len(X)-1)
```

```python
means=[np.mean(np.random.normal(loc=mu, scale=sigma,size=100)) for i in range(1000)] #sample means
```

```python
means[:10]
```

$E(\bar{X})=\mu$

$Var(\bar{X})=\sigma^2/N$

```python
np.mean(means)
```

```python
np.var(means) # 4/100
```

```python
sns.kdeplot(means)
```

2. Bernoulli distribution

```python
p=0.7
```

```python
np.random.binomial(n=1,p=p)
```

```python
X=np.random.binomial(n=1,p=p,size=100)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X).groupby(0)[0].count()
```

```python
np.mean(X)
```

```python
np.var(X)
```

```python
means=[np.mean(np.random.binomial(n=1,p=p,size=100)) for i in range(1000)]
```

```python
np.mean(means) # 0.7
```

```python
np.var(means)   # 0.7*0.3/100 =0.0021
```

```python
sns.kdeplot(means) # central limit theorem => N(mu, sigma^2 )
```
