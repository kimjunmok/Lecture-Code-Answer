1. Conduct a test whether the mean of the weights is 46kg. (Suppose the observations are a random sample from a normal population and we know that the population standard deviation is 3.5.)

   Hypothesis: 

   $H_0: \mu=46~~~$vs$~~~H_1:\mu\neq 46$

```python
# Test Statistic

Z=(np.mean(samp)-46)/3.5*np.sqrt(len(samp)) #~N(0,1) under H_0
```

```python
Z 
```

cdf: $\Phi(z)=P(Z\leq z)$

```python
# p-value

import scipy.stats as stats

p_val=2*(1-stats.norm.cdf(abs(Z)))
```

$e-06=10^{-6}$

```python
p_val 
```

```python
stats.norm.ppf(1-0.05/2) # significance level =0.05
```

$Z_{\frac{\alpha}{2}}$

$\alpha=0.05$-> 1.96

```python
np.abs(Z)>stats.norm.ppf(1-0.05/2)
```

2. Suppose population is not normal and the population variance is not known. What should we do?

```python
len(samp)
```

```python
# Test Statistic

Ssquare=np.var(samp)*len(samp)/(len(samp)-1)
S=np.sqrt(Ssquare)

T=(np.mean(samp)-46)/S*np.sqrt(len(samp))
```

```python
# p-value

p_val=2*(1-stats.norm.cdf(abs(T)))
```
