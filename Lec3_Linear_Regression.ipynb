# Linear Regression

## 1. Simple Linear Regression

```python
mydat=pd.read_csv('./data.csv', header=0)

mydat=mydat.dropna(axis=0)
```

```python
x = mydat[(mydat["gender"]==1) & (mydat["height"]!=0)]["height"]
y = mydat[(mydat["gender"]==1) & (mydat["height"]!=0)]["weight"]
plt.scatter(x, y,s=3.)
plt.xlabel("height")
plt.ylabel("weight")
plt.show()
```

### Estimation

$$\hat{\beta}_1=\frac{\sum_{i=1}^N(x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^N(x_i-\bar{x})^2}=\frac{s_{XY}}{s_{XX}}$$

$ $

$$\hat{\beta}_0=\bar{y}-\hat{\beta}_1\bar{x}$$

```python
x_c=x-np.mean(x)
y_c=y-np.mean(y)

s_xx=np.dot(x_c,x_c) # inner product 
s_xy=np.dot(x_c,y_c) 

beta_1=s_xy/s_xx
beta_0=np.mean(y)-beta_1*np.mean(x)
```

```python
beta_1, beta_0
```

```python
plt.scatter(x, y,s=3.)
plt.xlabel("height")
plt.ylabel("weight")

y_hat=beta_0+beta_1*x

plt.plot(x,y_hat,label='y='+str(np.round(beta_0,4))+'+'+str(np.round(beta_1,4))+'x',color='orange')
plt.legend()

plt.show()
```

```python
MSE=np.dot(y-y_hat,y-y_hat)/(len(x)-2)
```
